```python
#outlier treatment for Monetory,Frequency,Recency

Q1 = RFM['TotalAmount'].quantile(0.25)
Q3 = RFM['TotalAmount'].quantile(0.75)
IQR = Q3 - Q1
RFM = RFM[(RFM['TotalAmount'] >= Q1 - 1.5*IQR) & (RFM['TotalAmount'] <= Q3 + 1.5*IQR)]

```

```python

Q1 = RFM['ConsumptionFrequency'].quantile(0.25)
Q3 = RFM['ConsumptionFrequency'].quantile(0.75)
IQR = Q3 - Q1
RFM = RFM[(RFM['ConsumptionFrequency'] >= Q1 - 1.5*IQR) & (RFM['ConsumptionFrequency'] <= Q3 + 1.5*IQR)]


```

```python

Q1 = RFM['Recency'].quantile(0.25)
Q3 = RFM['Recency'].quantile(0.75)
IQR = Q3 - Q1
RFM = RFM[(RFM['Recency'] >= Q1 - 1.5*IQR) & (RFM['Recency'] <= Q3 + 1.5*IQR)]


```

## Affinity propagation

```python
rfm_numpy=RFM_norm2.to_numpy()
```

```python
rfm_numpy
```

```python
from sklearn.cluster import AffinityPropagation
from sklearn.metrics import silhouette_score
```

```python
sil=[]
pref=range(-500,0,100)
for i in pref:
    model_ap=AffinityPropagation(preference=i,verbose=True,damping=0.8)
    model_ap.fit(rfm_numpy)
    label=model_ap.labels_
    sil.append(silhouette_score(rfm_numpy,label))
plt.plot(pref,sil)
```

```python
import time
```

```python
ap=AffinityPropagation(damping=0.8,preference=-500)
start=time.time()
ap.fit(rfm_numpy)
end=time.time()
label=ap.labels_
print("No of cluster:{0}".format(max(label)+1))
print("Time:{0}".format(end-start))
```

```python
del ap
```

```python
ap=AffinityPropagation(damping=0.8,preference=-400)
start=time.time()
ap.fit(rfm_numpy)
end=time.time()
label=ap.labels_
print("No of cluster:{0}".format(max(label)+1))
print("Time:{0}".format(round((end-start),3)))
```

```python
def Threedplot(set1,colours='gray',title1='3Dplotting'):
    fig=plt.figure(figsize=(20,10))
    ax=plt.axes(projection='3d')
    ax.scatter(rfm_numpy[:,0],rfm_numpy[:,1],rfm_numpy[:,2],c=colours)
    ax.set_xlabel('TotalAmount',fontsize=20)
    ax.set_ylabel('ConsumptionFrequency',fontsize=20)
    ax.set_zlabel('Recency',fontsize=20)
    ax.set_title(title1,fontsize=30)
    fig.savefig('3DAffinityPropagation.png')
    plt.show()
```

```python
Threedplot(rfm_numpy,ap.labels_)
```

```python
def cluster_plots(set1,colours='gray',title1 = 'TotalAmount vs Consumption Frequency',title2='Consumption Frequency vs Recency'):
    fig,(ax1,ax2) = plt.subplots(1, 2,figsize=(15,9))
    ax1.set_title(title1,fontsize=14)
    ax1.set_xlim(min(set1[:,0]), max(set1[:,0]))
    ax1.set_ylim(min(set1[:,1]), max(set1[:,1]))
    ax1.scatter(set1[:, 0], set1[:, 1],s=20,lw=0,c=colours)
    ax2.set_title(title2,fontsize=14)
    ax2.set_xlim(min(set1[:,1]), max(set1[:,1]))
    ax2.set_ylim(min(set1[:,2]), max(set1[:,2]))
    ax2.scatter(set1[:, 1], set1[:, 2],s=20,lw=0,c=colours)
    fig.tight_layout()
    fig.savefig('AffinityClustering.png')
    plt.show()
    
```

```python
cluster_plots(rfm_numpy,ap.labels_)
```

```python
ap_cluster_centers=[[ 0.22554895,  0.9515217 , -0.69712878],
       [-0.60338054, -0.56476456,  1.48106233],
       [-0.4430602 , -0.54030833, -0.52209557],
       [ 1.88404439,  0.56022202, -0.66795658],
       [ 1.91851903,  2.17433321, -0.65823251]]
```

```python
np.array(ap_cluster_centers)
```

```python
km=KMeans(n_clusters=5,init=np.array(ap_cluster_centers))
```
